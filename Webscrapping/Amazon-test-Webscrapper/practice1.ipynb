```python
from bs4 import BeautifulSoup
from pip._vendor import requests
import pandas as pd
from IPython.display import display, HTML


def title(soup):
    title = soup.find("span", attrs={"id": "productTitle"})
    title = title.get_text(strip=True) if title else ""
    
    return title


def prices(soup):
    price = soup.find("span", attrs={"class": "a-price-whole"})
    price = price.get_text(strip=True) if price else ""
    
    return f"  €{price}00 "


def ratings(soup):
    rating = soup.find("span", attrs={"class": "a-icon-alt"})
    rating = rating.get_text(strip=True) if rating else ""
    
    return rating


URL = "https://www.amazon.co.uk/s?k=playstation+5&crid=1DZW6085PFENM&sprefix=playst%2Caps%2C76&ref=nb_sb_ss_ts-doa-p_2_6"

HEADERS = {
    "User-Agent": "Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/116.0.0.0 Safari/537.36"
}


response = requests.get(URL, headers=HEADERS)
content = BeautifulSoup(response.text, "html")
links = content.find_all("a", class_="a-link-normal s-underline-text s-underline-link-text s-link-style a-text-normal"
) 
    
PRODUCT_LIST = []

for link in links:
    path = link.get("href")
    if path.startswith("/"):
        product_path = "https://amazon.co.uk" + path
    else:
        product_path = path
        
    new_response = requests.get(product_path, headers=HEADERS)
    new_soup = BeautifulSoup(new_response.text, "html")
    
    products = {
        "Title": title(new_soup),
        "Prices": prices(new_soup),
        "Ratings": ratings(new_soup)
    }
    
    PRODUCT_LIST.append(products)
    


df = pd.DataFrame(PRODUCT_LIST)


display(HTML(df.to_html()))
```
